```python
# importing pandas library
import pandas as pd
```

```python
# creating a dataframe
df = pd.DataFrame([
    ('The Hobbit','J.R.R. Tolkien'),
    ('Robinson Crusoe','Daniel Defoe'),
    ('Moby-Dick','Herman Melville')
    ])
df
```

```python
# technique for renaming column labels is to call the rename method on the DataFrame object, then passing our list of label values to the columns parameter:
# renaming column in pandas 
df.rename(columns={0:'title',1:'author'},inplace=True)
df
```

```python
#We can modify the column labels by adding the following line:
df.columns = ['official_title','official_author']
df
```

```python
# reading the dataset
data_BM = pd.read_csv("bigmart_data.csv")
# drop the null values
data_BM = data_BM.dropna(how="any")
# view the top results
data_BM.head()
  
```

```python
# sorting dataframes
# pandas data frames has two useful functions
# sort_values(): to sort pandas data frame by one or more columns 
# sort_index(): to sort pandas data frame by row index
# each of these functions come with numerous options like sorting the data frame in 
# specific order(ascending or descending order),sorting in place,sorting with missing values ,sorting by specific algorithm etc 
# suppose we want to sort the data frame by "Outlet_Establishment_Year" then we will use sort_values

# sort by year
sorted_data = data_BM.sort_values(by='Outlet_Establishment_Year')
#print sorted_data
sorted_data[:5]
```

```python
# sort in place and descending order
data_BM.sort_values(by='Outlet_Establishment_Year', ascending=False, inplace=True)
data_BM[:5]
```

```python
# reading the dataset
data_BM = pd.read_csv("bigmart_data.csv")
# drop the null values
data_BM = data_BM.dropna(how="any")
# sort by multiple columns 
data_BM.sort_values(by=['Outlet_Establishment_Year','Item_Outlet_Sales'], ascending=False)[:5]
```

```python
# changed the order of columns
data_BM.sort_values(by=['Item_Outlet_Sales','Outlet_Establishment_Year'], ascending=False, inplace=True)
data_BM[:5]
```

```python
# sort data frame by index
data_BM.sort_index(inplace=True)
data_BM
```
