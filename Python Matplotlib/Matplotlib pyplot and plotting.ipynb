# Matplotlib Pyplot
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
# Now the Pyplot package can be referred to as plt.
import matplotlib.pyplot as plt
```

Example:
Draw a line in a diagram from position (0,0) to position (10,300):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,10])
ypoints = np.array([0,300])

plt.plot(xpoints,ypoints)
plt.show()
```

# Matplotlib Plotting

Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.
By default, the plot() function draws a line from point to point.
The function takes parameters for specifying points in the diagram.
Parameter 1 is an array containing the points on the x-axis.
Parameter 2 is an array containing the points on the y-axis.

for instance; If we need to plot a line from (1, 6) to (7, 20), we have to pass two arrays [1, 7] and [6, 20] to the plot function.


```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,7])
y_axis_points = np.array([6,20])

plt.plot(x_axis_points, y_axis_points)
plt.show()
```

# Plotting Without Line
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

Example:
Draw two points in the diagram, one at position (1, 6) and one in position (7, 20):

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,7])
y_axis_points = np.array([6,20])

plt.plot(x_axis_points, y_axis_points, 'o')
plt.show()
```

# Multiple Points
You can plot as many points as you like, just make sure you have the same number of points in both axis.

Example:
Draw a line in a diagram from position (1, 6) to (3, 14) then to (5, 4) and finally to position (7, 20):

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points)
plt.show()
```

# Default X-Points
If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, etc. depending on the length of the y-points.

Example:
Plotting without x_axis_points:

```python
# So, if we take the same example as above, and leave out the x-points, the diagram will look like this:
import matplotlib.pyplot as plt
import numpy as np

y_axis_points = np.array([6,14,4,20,13,15])

plt.plot(y_axis_points)
plt.show()
```

The x_axis_points in the example above is [0, 1, 2, 3, 4, 5].
