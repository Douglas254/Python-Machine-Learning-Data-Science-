# Matplotlib Markers


Markers

You can use the keyword argument `marker` to emphasize each point with a specified marker:

Example:
Mark each point with a circle:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, marker='o')
plt.show()
```

Example:
    
Mark each point with a star:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, marker='*')
plt.show()
```

# Format Strings fmt
You can use the shortcut string notation parameter to specify the marker.

This parameter is also called `fmt`, and is written with this syntax:

`marker|line|color`

```python
# Example
# Mark each point with a circle:

import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o:r')
plt.show()
```

# Marker Size
You can use the keyword argument `markersize` or the shorter version, `ms` to set the size of the markers:

Example:
    
Set the size of the markers to 20:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o:b', ms=20)
plt.show()
```

# Marker Color
You can use the keyword argument `markeredgecolor` or the shorter `mec` to set the color of the edge of the markers:

Example:
    
Set the EDGE color to red:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o:g', ms=20, mec = 'r')
plt.show()
```

You can use the keyword argument `markerfacecolor` or the shorter `mfc` to set the color inside the edge of the markers:

Example:
    
Set the FACE color to white:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o:g', ms=20, mec = 'r',mfc='w')
plt.show()
```

Use both the `mec` and `mfc` arguments to color of the entire marker:

Example:
    
Set the color of both the edge and the face to red:

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o-.g', ms=20, mec = 'r', mfc='r')
plt.show()
```

You can also use Hexadecimal color values instead



```python
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o--k', ms=20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
# using color names
import matplotlib.pyplot as plt
import numpy as np

x_axis_points = np.array([1,3,5,7])
y_axis_points = np.array([6,14,4,20])

plt.plot(x_axis_points,y_axis_points, 'o-k', ms=20, mec = 'purple', mfc = 'purple')
plt.show()
```
