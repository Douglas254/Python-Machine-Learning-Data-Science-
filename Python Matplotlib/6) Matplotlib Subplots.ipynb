# Matplotlib Subplots
<b>Display Multiple Plots</b>

With the `subplots()` function you can draw multiple plots in one figure:

Example:

Draw 2 plots:

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([3,8,1,8,10])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)


# Plot 2:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([5,10,20,30,45])

plt.subplot(1,2,2)
plt.plot(x_axis, y_axis)

plt.show()
````

# The subplots() Function
The `subplots()` function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.```

```plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.```
So, if we want a figure with 2 rows and 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

Example:

Draw 2 plots on top of each other:

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([3,8,1,8,10])

plt.subplot(2,1,1)
plt.plot(x_axis, y_axis)


# Plot 2:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([5,10,20,30,45])

plt.subplot(2,1,2)
plt.plot(x_axis, y_axis)

plt.show()
````

<b>You can draw as many plots you like on one figure, just descibe the number of `rows`, `columns`, and the `index` of the plot.</b>

Example:

Draw 6 plots:

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([3, 8, 1, 15, 10])

plt.subplot(2, 3, 1)
plt.plot(x_axis, y_axis)


# Plot 2:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([5, 10, 20, 30, 45])

plt.subplot(2, 3, 2)
plt.plot(x_axis, y_axis)


# Plot 3:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([3, 8, 1, 15, 10])

plt.subplot(2, 3, 3)
plt.plot(x_axis, y_axis)


# Plot 4:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([5, 10, 20, 30, 45])

plt.subplot(2, 3, 4)
plt.plot(x_axis, y_axis)


# Plot 5:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([3, 8, 1, 15, 10])

plt.subplot(2, 3, 5)
plt.plot(x_axis, y_axis)


# Plot 6:
x_axis = np.array([0, 1, 2, 3, 4])
y_axis = np.array([5, 10, 20, 30, 45])

plt.subplot(2, 3, 6)
plt.plot(x_axis, y_axis)


plt.show()
````

# Title
You can add a title to each plot with the `title()` function:

Example:
    
2 plots, with titles:

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([3,8,1,8,10])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)
plt.title("SALES")

# Plot 2:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([5,10,20,30,45])

plt.subplot(1,2,2)
plt.plot(x_axis, y_axis)
plt.title("INCOME")

plt.show()
````

# Super Title
You can add a title to the entire figure with the `suptitle()` function:

Example:
    
Add a title for the entire figure:

````python
import numpy as np
import matplotlib.pyplot as plt

# Plot 1:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([3,8,1,8,10])

plt.subplot(1,2,1)
plt.plot(x_axis, y_axis)
plt.title("SALES")

# Plot 2:
x_axis = np.array([0,1,2,3,4])
y_axis = np.array([5,10,20,30,45])

plt.subplot(1,2,2)
plt.plot(x_axis, y_axis)
plt.title("INCOME")

plt.suptitle("FREDOUTECH BUSINESS")
plt.show()
````
