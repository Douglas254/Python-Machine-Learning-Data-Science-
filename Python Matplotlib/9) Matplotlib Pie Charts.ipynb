# Matplotlib Pie Charts
Creating Pie Charts
With Pyplot, you can use the `pie()` function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15]) 

plt.pie(n)
plt.show()
```

As you can see the pie chart draws one piece (called a wedge) for each value in the array (in this case [35, 25, 25, 15]).

By default the plotting of the first wedge starts from the x-axis and move counterclockwise



    

<b>Note:</b> The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: `x/sum(x)`

# Labels
Add labels to the pie chart with the `labels` parameter.

The label parameter must be an array with one `labels` for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]

plt.pie(n, labels=array_labels)
plt.show()
```

# Start Angle
As mentioned the default start angle is at the <b>x-axis</b>, but you can change the start angle by specifying a `startangle` parameter.

The `startangle` parameter is defined with an angle in degrees, default angle is <b>0</b>




```python
# Start the first wedge at 90 degrees

import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]

plt.pie(n, labels=array_labels, startangle = 90)
plt.show() 
```

# Explode
Maybe you want one of the wedges to stand out? The `explode` parameter allows you to do that

The `explode` parameter, if specified, and not `None`, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
# Example
# Pull the "Oranges" wedge 0.3 from the center of the pie

import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]
my_explode = [0.3, 0, 0, 0]

plt.pie(n, labels=array_labels, explode=my_explode)
plt.show() 
```

# Shadow
Add a shadow to the pie chart by setting the shadows parameter to `True`

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]
my_explode = [0.3, 0, 0, 0]

plt.pie(n, labels=array_labels, explode=my_explode, shadow=True)
plt.show()
```

# Colors
You can set the color of each wedge with the `colors` parameter.

The `colors` parameter, if specified, must be an array with one value for each wedge

```python
# Specifying a new color for each wedge

import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]
my_colors = ["orange", "g", "darkblue", "#fbff0a"]


plt.pie(n, labels=array_labels, colors=my_colors)
plt.show()
```

# Legend
To add a list of explanation for each wedge, use the `legend()` function:

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]
my_colors = ["orange", "g", "darkblue", "#fbff0a"]


plt.pie(n, labels=array_labels, colors=my_colors)
plt.legend()
plt.show()
```

# Legend With Header
To add a header to the legend, add the `title` parameter to the `legend` function.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([35,25,25,15])
array_labels = ["Oranges","Avocado","Grapes","Banana"]
my_colors = ["orange", "g", "darkblue", "#fbff0a"]


plt.pie(n, labels=array_labels, colors=my_colors)
plt.legend(title='FAVOURITE FRUITS')
plt.show()
```
